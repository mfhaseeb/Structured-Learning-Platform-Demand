```python
#For Question4
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

ratings = [
    4, 3, 5, 4, 4, 5, 3, 5, 4, 4, 3, 3, 3, 5, 4, 1,
    3, 5, 3, 2, 3, 5, 2, 4, 4, 4, 4, 3, 4, 3, 2, 5,
    4, 5, 1, 2, 3, 2, 4, 1, 5, 3, 3, 4, 3, 3, 4, 3,
    1, 3, 3, 4, 1, 5, 5, 4, 4, 5]


# === Descriptive Stats ===
mean_rating = np.mean(ratings)
median_rating = np.median(ratings)
mode_rating = stats.mode(ratings, keepdims=True).mode[0]

print("Mean:", mean_rating)
print("Median:", median_rating)
print("Mode:", mode_rating)

# === Wilcoxon Signed-Rank Test (against median = 3) ===
diffs = np.array(ratings) - 3
wilcoxon_result = stats.wilcoxon(diffs[diffs != 0], alternative='two-sided')
print("\nWilcoxon Test Statistic (W):", wilcoxon_result.statistic)
print("p-value:", wilcoxon_result.pvalue)

# === Distribution Plot ===
df = pd.DataFrame({'Rating': ratings})
plt.figure(figsize=(8, 5))
sns.countplot(x='Rating', data=df, palette='viridis')
plt.title('Distribution of Student Ratings (1 = Very Poor, 5 = Excellent)', fontsize=14)
plt.xlabel('Rating')
plt.ylabel('Number of Students')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("rating_distribution.png", dpi=300)  # Save the plot for your report
plt.show()
```

```python
# Q ~ Q4 + Q8

import pandas as pd
from statsmodels.miscmodels.ordinal_model import OrderedModel
import seaborn as sns
import matplotlib.pyplot as plt

# Your data
Q7 = [
    3, 2, 4, 5, 3, 5, 3, 4, 3, 3, 2, 2, 2, 
    5, 4, 1, 2, 4, 2, 2, 4, 3, 2, 3, 5, 2, 
    4, 2, 4, 2, 2, 4, 3, 3, 4, 4, 3, 4, 4, 5, 
    2, 4, 4, 1, 4, 3, 4, 4, 4, 1, 3, 2, 1, 1, 4, 3, 1, 4
]

Q4 = [
    4, 3, 5, 4, 4, 5, 3, 5, 4, 4, 3, 3, 3, 5, 4, 1,
    3, 5, 3, 2, 3, 5, 2, 4, 4, 4, 4, 3, 4, 3, 2, 5,
    4, 5, 1, 2, 3, 2, 4, 1, 5, 3, 3, 4, 3, 3, 4, 3,
    1, 3, 3, 4, 1, 5, 5, 4, 4, 5
]

# Simulated dependent variable Q8
Q8 = [5 if (q4 < 3 or q7 < 3) else 3 if (q4 == 3 or q7 == 3) else 1 for q4, q7 in zip(Q4, Q7)]

# DataFrame
data = pd.DataFrame({'Q4': Q4, 'Q7': Q7, 'Q8': Q8})

# Fit the model
model = OrderedModel(data['Q8'], data[['Q4', 'Q7']], distr='logit')
result = model.fit(method='bfgs', disp=False)

# Create summary table
summary_df = pd.DataFrame({
    'Coefficient': result.params,
    'Std. Error': result.bse,
    'z-value': result.tvalues,
    'p-value': result.pvalues
}).round(4)

summary_df['CI Lower'] = result.conf_int()[0].round(4)
summary_df['CI Upper'] = result.conf_int()[1].round(4)

summary_df.index.name = 'Variable'
summary_df.reset_index(inplace=True)

# Style the summary
styled_summary = summary_df.style\
    .background_gradient(subset=['p-value'], cmap='RdYlGn_r', low=0, high=1)\
    .bar(subset=['Coefficient'], color='#5fba7d')\
    .format({
        "Coefficient": "{:.4f}",
        "Std. Error": "{:.4f}",
        "z-value": "{:.2f}",
        "p-value": "{:.4f}",
        "CI Lower": "{:.4f}",
        "CI Upper": "{:.4f}"
    })\
    .set_caption(" Ordered Logistic Regression Summary (Q8 ~ Q4 + Q7)")

styled_summary
```
